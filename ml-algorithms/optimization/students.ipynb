```python
import optimization
```

```python
# Double-occupancy rooms
dorms=['Zeus','Athena','Hercules','Bacchus','Pluto']
print(dorms)
```

```python
# Students and their two room preferences each
prefs=[('Toby',('Bacchus','Hercules')),
    ('Steve',('Zeus','Pluto')),
    ('Andrea',('Athena','Zeus')),
    ('Sarah',('Zeus','Pluto')),
    ('Dave',('Athena','Bacchus')),
    ('Jeff',('Hercules','Pluto')),
    ('Fred',('Pluto','Athena')),
    ('Suzie',('Bacchus','Hercules')),
    ('Laura',('Bacchus','Hercules')),
    ('Neil',('Hercules', 'Athena'))]
print(prefs)
```

```python
domain=[(0,(len(dorms)*2)-i-1) for i in range(0, len(dorms)*2)]
print(domain)
```

```python
def printsolution(vec):
    slots=[]
    # Create two slots per room
    for i in range(len(dorms)):
        slots+=[i,i]
    # Loop to assign students to rooms
    for i in range(len(vec)):
        x=int(vec[i])
        # Pick any slot from the remaining ones
        dorm=dorms[slots[x]]
        # Print student name and the room they were assigned to
        print(prefs[i][0],dorm)
        # Remove this slot
        del slots[x]
```

```python
printsolution([0,0,0,0,0,0,0,0,0])
```

```python
def dormcost(vec):
    cost=0
    # Create list of slots
    slots=[0,0,1,1,2,2,3,3,4,4]
    # Loop over students
    for i in range(len(vec)):
        x=int(vec[i])
        dorm=dorms[int(slots[x])]
        pref=prefs[i][1]
        # Primary preference costs 0, secondary costs 1
        if pref[0]==dorm:
            cost+=0
        elif pref[1]==dorm:
            cost+=1
        else:
            cost+=3
        # If the room is not in the preference list, cost increases by 3
        del slots[x]
    return cost
```

```python
s = optimization.randomoptimize(domain, dormcost)
dormcost(s)
```

```python
s = optimization.geneticoptimize(domain, dormcost)
dormcost(s)
print(s)
```

```python
printsolution(s)
```

```python
s = optimization.hillclimb(domain,dormcost)
```

```python
dormcost(s)
```
