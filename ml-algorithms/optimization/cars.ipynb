```python
import optimization
```

```python
cars = ['Mercedes','ferrari','lambo', 'traktor', 'uaz', 'zaz', 'gaz','vaz']
```

```python
wastes = {('Mercedes'): [16000, 3000, 65000, 1444400, 25000, 64500], 
          ('ferrari'): [16000, 3000, 65000, 25000, 25000, 64500],
         ('lambo'): [16000, 3000, 3000, 1444400, 25000, 64500],
         ('traktor'): [25000, 3000, 65000, 1444400, 25000, 64500],
         ('uaz'): [16000, 3000, 65000, 1444400, 3000, 64500],
         ('zaz'): [16000, 65000, 3000, 3000, 3000, 64500],
         ('gaz'): [16000, 3000, 65000, 1444400, 25000, 64500],
         ('vaz'): [16000, 3000, 65000, 3000, 25000, 64500]}
print(wastes)
#cars = [('Mercedes', 16000, 3000, 65000, 1444400, 25000, 64500), ('ferrari', 16000, 3000, 65000, 1444400, 25000, 64500)]
#print(cars)
```

```python
def printresult(r):
    print('%10s %10s %10s %10s %10s %10s %10s'%('Model', 'KASKO', 'OSAGO','Fuel cost', 'Resale val', 'Tax', 'Service'))
    model = cars[int(r[0])]
    out = wastes[(model)]
    print('%10s %10s %10s %10s %10s %10s %10s'%(model, out[0], out[1], out[2], out[3], out[4], out[5]))
```

```python
s = [1]
printresult(s)
```

```python
def ownershipcost(sol):
    totalwaste = 0
    model = cars[int(sol[0])]
    for v in range(5):
        totalwaste += wastes[(model)][v]
        
    return totalwaste
```

```python
domain = [(0,8)]
```

```python
ownershipcost(s)
```

```python
printresult(s)
```

```python
s=optimization.randomoptimize(domain, ownershipcost)
print(s)
```
