```python
import random
import math
from PIL import Image, ImageDraw
import optimization
```

```python
people=['Charlie','Augustus','Veruca','Violet','Mike','Joe','Willy',
'Miranda']
print(people)
print(len(people))
```

```python
links=[('Augustus','Willy'),
('Mike','Joe'),
('Miranda','Mike'),
('Violet','Augustus'),
('Miranda','Willy'),
('Charlie','Mike'),
('Veruca','Joe'),
('Miranda','Augustus'),
('Willy','Augustus'),
('Joe','Charlie'),
('Veruca','Augustus'),
('Miranda','Joe')]
print(links)
```

```python
domain=[(10,370)]*(len(people)*2)
print(domain)
```

```python
def crosscount(v):
    # Convert list of numbers into a dict of person: (x, y) coordinates
    loc=dict([(people[i],(v[i*2],v[i*2+1])) for i in range(0,len(people))])
    total=0
    # Iterate over all pairs of edges
    for i in range(len(links)):
        for j in range(i+1,len(links)):
            # Get the endpoint coordinates
            (x1,y1),(x2,y2)=loc[links[i][0]],loc[links[i][1]]
            (x3,y3),(x4,y4)=loc[links[j][0]],loc[links[j][1]]
            den=(y4-y3)*(x2-x1)-(x4-x3)*(y2-y1)
            # den == 0 means the lines are parallel
            if den==0:
                continue
            # Otherwise ua and ub are the fractions along each segment
            # where the intersection point lies
            ua=((x4-x3)*(y1-y3)-(y4-y3)*(x1-x3))/den
            ub=((x2-x1)*(y1-y3)-(y2-y1)*(x1-x3))/den
            # If for both segments the fraction is between 0 and 1,
            # then they intersect
            if ua>0 and ua<1 and ub>0 and ub<1:
                total+=1
    return total
```

```python
def drawnetwork(sol):
    # Create image
    img=Image.new('RGB',(400,400),(255,255,255))
    draw=ImageDraw.Draw(img)
    # Create position dictionary
    pos=dict([(people[i],(sol[i*2],sol[i*2+1])) for i in range(0,len(people))])
    # Draw connecting lines
    for (a,b) in links:
        draw.line((pos[a],pos[b]),fill=(255,0,0))
    # Draw people's names
    for n,p in pos.items():
        draw.text(p,n,(0,0,0))
    img.show()
```

```python
sol = optimization.randomoptimize(domain, crosscount)
```

```python
crosscount(sol)
```

```python
drawnetwork(sol)
```
